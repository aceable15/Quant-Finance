# Question One

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import pandas as pd
```

```python
def f_system(x):
    y=float(x[0])
    y_star=float(x[1])
    lamda=float(x[2])
    return np.array([100/95.46-1-y, (100*(1-lamda)+0.52*100*lamda)/94.5-1-y_star,(1-(1+y)/(1+y_star))/(1-0.52)-lamda])

sol = fsolve(f_system, [1,1,1])
print(f'y is {sol[0]:.2f}, y_star is {sol[1]:.2f}, lambda is {sol[2]:.2f}')
```
