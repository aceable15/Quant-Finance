Portfolio's Elements:

* Capital: The amount of money we have available to invest or speculate.
* Instruments: These are the assets available for inclusion in our portfolio management strategy.
* Volatility: Also called asset risk, indicates the movement of the asset for the period analysed.
* Cost: Amount of money needed to buy/sell-short an asset.
* Position weight: The amount of capital we allocate to each asset and/or strategy.
* Return: Absolute returns are the return of our portfolio and relative returns are the return of our portfolio compared to a benchmark. When we are not comparing our returns with anybody, absolute return is a good measure, but when we need to compare performances we use the relative return.
* Risk or Volatility: This is the (estimated) amount that the portfolio assumes.

### Rebalance(Model)

Here is the rebalance model for testing the performance in last 4 quarters

```python
# Price of all stock instruments at the beginning of each period, we took the last 4 quarters
price = np.array(
    [
        [180.96,170.40,399.93,338.54],
        [173.52,163.56,361.53,224.85],
        [138.51,109.56,227.26,160.03],
        [142.21,115.88,242.40,138.61],
    ]
)
quantity = np.zeros_like(price)
# We assume to invest 100 dollar
initial_investment = 100
target_weight = np.array([0.25, 0.25, 0.25, 0.25])
for i in range(price.shape[0]):
    if i == 0:
        quantity[i] = initial_investment * target_weight / price[i]
    else:
        portfolio_value = (quantity[i-1] * price[i]).sum()
        quantity[i] = portfolio_value * target_weight / price[i]
columns = pd.MultiIndex.from_product([["price", "quantity"],list("AATM")])
df8 = pd.DataFrame(np.hstack([price, quantity]), columns=columns)
df8["portfolio_value"] = (df8["price"] * df8["quantity"]).sum(axis=1)
print(df8)
```

```python
# Visualize our portfolio_value
fig = plt.figure(4)
dates = ['2022-01-01','2022-4-1', '2022-7-1', '2022-10-1']
date = [datetime.strptime(d, '%Y-%m-%d').date() for d in dates]
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y/%m/%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.xticks(pd.date_range('2022-1-1','2022-11-24',freq='90d'))
plt.plot(date, df8['portfolio_value'],label='portfolio value') 
plt.xlabel ('Dates') 
plt.ylabel ('Portfolio Value')
plt.title('Portfolio performance in last four quarters')
plt.legend()
plt.grid(True)
plt.show()
```

```python
for stock, weight in zip([aapl, meta, tsla, amzn],[0.25, 0.25, 0.25, 0.25]):
    stock['weighted daily return']=stock['normalized_price']*weight
aapl.head()
```

```python
total_stock=pd.concat([aapl['weighted daily return'], meta['weighted daily return'], tsla['weighted daily return'], amzn['weighted daily return']],axis=1)
total_stock.columns=['aapl', 'meta', 'tsla', 'amzn']
total_stock.head()
```

```python
#assume we invest 10000
total_invest=total_stock*10000
total_invest.head()
```

```python
total_invest['Total Pos']=total_invest.sum(axis=1)
total_invest.head()
```

```python
#Total Portfolio Value
plt.style.use('ggplot')
total_invest['Total Pos'].plot(label='Total Pos')
plt.legend(loc='best')
plt.title('Total Portfolio Value')
```

```python
total_invest.drop('Total Pos',axis=1).plot(figsize=(8,4))
plt.title('Each stock Value')
```

```python
#Daily Return
total_invest['daily return']=total_invest['Total Pos'].pct_change(1)
total_invest
```

```python
total_invest['cumulret'] = (total_invest['daily return']+1).cumprod()-1
total_invest
```

```python
#Cumulative Return
cumulative_return=total_invest['Total Pos'].iloc[-1]/total_invest['Total Pos'].iloc[0]-1
print(cumulative_return)
```

```python
total_invest['logret']=np.log(1+total_invest['daily return'])
total_invest['cumullogret']=total_invest['logret'].cumsum()
total_invest['other_cumulret']=np.exp(total_invest['cumullogret'])-1
total_invest
```
